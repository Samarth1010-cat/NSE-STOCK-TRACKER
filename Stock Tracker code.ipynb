```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta

# 1. Define 10 top NSE stock tickers across major sectors
tickers = [
    'RELIANCE.NS',     # Energy / Oil & Gas
    'TCS.NS',          # IT Services
    'HDFCBANK.NS',     # Banking & Finance
    'INFY.NS',         # IT Services
    'ITC.NS',          # FMCG / Consumer Goods
    'ICICIBANK.NS',    # Banking & Finance
    'BHARTIARTL.NS',   # Telecom
    'SBIN.NS',         # State Bank (Banking)
    'HINDUNILVR.NS',   # FMCG / Consumer Goods
    'LT.NS'            # Engineering & Construction
]

# 2. Set the timeframe (Last 2 years from today)
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.today() - timedelta(days=730)).strftime('%Y-%m-%d')

print(f"🚀 Starting data extraction from {start_date} to {end_date}...\n")

all_stock_data = []

# 3. Download and process each stock
for ticker in tickers:
    print(f"📥 Downloading: {ticker}")

    # Fetch historical data from Yahoo Finance
    df = yf.download(ticker, start=start_date, end=end_date)

    # Reset index so Date becomes a standard workable column
    df = df.reset_index()

    # Fix multi-layered column headers if they occur
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    # Create a clean stock name indicator (remove '.NS')
    df['Stock_Name'] = ticker.replace('.NS', '')

    # 4. Metric Engineering (Calculations)
    # Calculate Daily Return %
    df['Daily_Return_Pct'] = ((df['Close'] - df['Open']) / df['Open']) * 100

    # Calculate 50-day and 200-day Simple Moving Averages (SMA)
    df['50_Day_SMA'] = df['Close'].rolling(window=50).mean()
    df['200_Day_SMA'] = df['Close'].rolling(window=200).mean()

    # Select and organize the exact columns we need
    essential_columns = [
        'Date', 'Stock_Name', 'Open', 'High', 'Low',
        'Close', 'Volume', 'Daily_Return_Pct', '50_Day_SMA', '200_Day_SMA'
    ]
    df = df[essential_columns]

    # Add to our master list
    all_stock_data.append(df)

# 5. Combine all 10 stocks into a single big table
final_dataset = pd.concat(all_stock_data, ignore_index=True)

# Drop early rows where moving averages can't calculate yet
final_dataset = final_dataset.dropna(subset=['Open', 'High', 'Low', 'Close'])

# 6. Save the data to a CSV file inside Google Colab
output_filename = "nse_10_stock_tracker_data.csv"
final_dataset.to_csv(output_filename, index=False)

print("\n=======================================================")
print(f"✅ SUCCESS! Cleaned dataset saved as: {output_filename}")
print(f"📈 Total rows generated for portfolio analysis: {len(final_dataset)}")
print("=======================================================")
```
